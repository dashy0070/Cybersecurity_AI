### Challenges in Detecting New, Unseen Phishing Strategies and Feature Engineering for Dynamic URLs

a. **Highlight the challenges in detecting new, unseen phishing strategies.**

Detecting new, unseen phishing strategies, often referred to as zero-day phishing attacks, poses significant challenges due to several factors:

1.  **Lack of Prior Knowledge/Signatures:** Traditional phishing detection systems heavily rely on blacklists, known phishing URLs, and signature-based detection. For novel attacks, these systems are ineffective because no prior signatures or malicious patterns exist in their databases.

2.  **Evolving Tactics:** Phishers constantly evolve their techniques to bypass existing defenses. They use sophisticated methods like polymorphic URLs, homoglyph attacks, dynamic content generation, and evasion techniques to make their attacks appear legitimate or circumvent detection mechanisms.

3.  **Mimicry and Obfuscation:** New phishing pages often mimic legitimate websites almost perfectly, making it difficult for both humans and automated systems to distinguish between them. They might use subtle visual cues, legitimate-looking domain names, and advanced obfuscation techniques to hide malicious code or redirect users.

4.  **Social Engineering Sophistication:** Phishing campaigns increasingly leverage advanced social engineering tactics, tailoring messages to specific individuals or organizations (spear phishing, whale phishing). These personalized attacks are harder to detect through generic filters as they often exploit trust and context.

5.  **Dynamic Content and Infrastructure:** Attackers frequently change their hosting infrastructure, use legitimate but compromised websites, or generate dynamic content to evade detection. This rapid mutation makes it hard to block them effectively.

6.  **Human Factor:** Ultimately, phishing exploits human vulnerabilities. Even with advanced technical defenses, a well-crafted, new phishing attempt can still trick users who are not aware of the latest threats or are under pressure.

7.  **Resource Constraints:** Developing and deploying real-time detection systems capable of identifying novel threats requires significant computational resources, advanced machine learning models, and continuous updates.

b. **Discuss feature engineering for dynamic URLs.**

Dynamic URLs often contain parameters that change based on user input, session IDs, or other variables. Feature engineering for dynamic URLs is crucial for distinguishing legitimate dynamic content from malicious ones, as phishing URLs often hide their intent within these dynamic parts. Here are some key approaches to feature engineering for dynamic URLs:

1.  **URL Length:** Phishing URLs are often longer due to attempts to embed keywords, multiple subdomains, or obfuscated parameters. The total length of the URL can be a simple yet effective feature.

2.  **Number of Parameters:** Count the number of `key=value` pairs in the query string. Malicious URLs might have an unusually high or low number of parameters.

3.  **Entropy of Parameters:** Measure the randomness or complexity of the parameter values. Highly entropic or random-looking strings within parameters could indicate obfuscation or malicious intent (e.g., encoded payloads).

4.  **Presence of Specific Keywords/Characters in Parameters:** Look for keywords like `login`, `verify`, `account`, or unusual characters (`%`, `$`, `&`, `#`) within parameter names or values that might indicate an attempt to trick users or inject malicious code.

5.  **Parameter Value Type and Length:** Analyze the type of data expected in parameters (e.g., numbers, strings, dates) versus what is actually present. Extremely long or overly complex parameter values can be suspicious.

6.  **Domain vs. Parameter Domain Discrepancy:** In some phishing attacks, the URL might appear legitimate in the main domain, but a crucial parameter might contain a different, malicious domain or IP address. Extracting and comparing these can be a strong indicator.

7.  **Path Components Analysis:** Similar to parameters, analyze individual components of the URL path (separated by `/`). The number of path components, their length, and the presence of unusual characters or keywords within them can be indicative.

8.  **Redirection Indicators:** Dynamic URLs are often used to facilitate redirects. Features like the presence of `redirect=`, `url=`, or `dest=` parameters, especially when combined with external domains, can signal malicious intent.

9.  **Subdomain and TLD Analysis:** While not strictly part of the dynamic parameters, the structure of subdomains within a dynamic URL can be engineered. For example, `login.example.com.malicious.com` uses multiple subdomains to confuse users.

10. **Encoding Detection:** Check for multiple layers of URL encoding (`%2520` instead of `%20`). Phishers often use double or triple encoding to hide malicious payloads.

Effective feature engineering for dynamic URLs requires a deep understanding of how attackers craft these URLs and a combination of linguistic, structural, and statistical analysis techniques.

### Data exploration and Cleaning

Here we will examin the dataset to understand the structure and properties.

Prompt:Preprocess the phishing dataset by converting labels (-1 = phishing, 1 = legitimate) into binary values (1 = phishing, 0 = legitimate) and check for missing values.


```python
import pandas as pd
from ucimlrepo import fetch_ucirepo

# Re-fetch the dataset to ensure 'y' is in its original state for transformation
# This ensures that if the cell is run multiple times, y is always correctly reset.
phishing_websites_temp = fetch_ucirepo(id=327)
y = phishing_websites_temp.data.targets.copy() # Use .copy() to avoid SettingWithCopyWarning initially

# Convert labels: -1 (phishing) to 1, 1 (legitimate) to 0
y['result'] = y['result'].replace({-1: 1, 1: 0})

print("\nDistribution of the label column (y) after conversion:")
print(y.value_counts())

# Check for missing values in features (X)
print("\nMissing values in features (X):")
print(X.isnull().sum())

# Check for missing values in targets (y)
print("\nMissing values in targets (y):")
print(y.isnull().sum())
```

Inference: preprocess the target variable y by converting the labels: -1 (phishing) will become 1, and 1 (legitimate) will become 0.

###  Feature Scaling

Prompt: Apply StandardScaler to normalize all numerical features except the label column.


```python
from sklearn.preprocessing import StandardScaler

# Initialize StandardScaler
scaler = StandardScaler()

# Apply StandardScaler to the features (X)
# The fit_transform method fits the scaler to X and then transforms X.
X_scaled = scaler.fit_transform(X)

# Convert the scaled array back to a DataFrame, preserving column names
X = pd.DataFrame(X_scaled, columns=X.columns)

print("Displaying the first 5 rows of features (X) after StandardScaler:")
display(X.head())
```

Inference: All numerical features in X were successfully standardized using StandardScaler, resulting in values centered around mean = 0 with standard deviation = 1.

### Train-Test Split

Prompt: Split the dataset into training and testing sets using an 80/20 stratified split.


```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets (80/20 split)
# stratify=y ensures that the proportion of target classes is the same in both train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)

print("\nDistribution of y_train:")
print(y_train.value_counts(normalize=True))

print("\nDistribution of y_test:")
print(y_test.value_counts(normalize=True))
```

Inference: The dataset was successfully split into 80% training and 20% testing sets, ensuring sufficient data for model training while reserving unseen data for evaluation.

### Build Deep Learning Model

Prompt: Build a deep neural network for phishing detection using TensorFlow/Keras with layers (128 → 128 → 64 → 1), ReLU activation, dropout regularization, and sigmoid output.


```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout

# Get the number of features from X_train
input_shape = X_train.shape[1]

# Build the Sequential model
model = Sequential([
    Dense(128, activation='relu', input_shape=(input_shape,)),
    Dropout(0.3), # Added dropout for regularization
    Dense(128, activation='relu'),
    Dropout(0.3), # Added dropout for regularization
    Dense(64, activation='relu'),
    Dropout(0.3), # Added dropout for regularization
    Dense(1, activation='sigmoid') # Output layer for binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Display model summary
print(model.summary())
```

Inference: A deep neural network with three hidden layers (128, 128, 64 neurons) and dropout regularization was successfully built to learn complex patterns in the dataset while reducing overfitting.

### Train Model


Prompt: Train the model using Adam optimizer, binary crossentropy loss, batch size 32, and 20 epochs. Also use class weights to handle imbalance.


```python
from sklearn.utils import class_weight
import numpy as np

# Convert y_train to a 1D numpy array for class_weight calculation
y_train_1d = y_train.values.flatten()

# Calculate class weights
# This assigns a higher weight to the minority class
class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train_1d),
    y=y_train_1d
)

# Convert class_weights to a dictionary for Keras
class_weight_dict = dict(enumerate(class_weights))
print(f"Calculated Class Weights: {class_weight_dict}")

# Ensure y_train and y_test are in the correct format for Keras (NumPy array)
y_train_np = y_train.values
y_test_np = y_test.values

# Train the model
history = model.fit(
    X_train,
    y_train_np,
    epochs=20,
    batch_size=32,
    validation_data=(X_test, y_test_np),
    class_weight=class_weight_dict,
    verbose=1
)

print("\nModel training complete.")
```

Inference: The deep learning model was successfully trained for 20 epochs, showing a steady increase in accuracy and decrease in loss, indicating effective learning and good convergence.

The final validation accuracy of 96.97% with low validation loss (0.0799).

###  Evaluate Model

Prompt: Evaluate the phishing detection model using accuracy, precision, recall, F1-score, confusion matrix, and ROC-AUC.

- Expected metrics:

- Confusion matrix

- Classification report

- ROC curve

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Make predictions on the test set
y_pred_proba = model.predict(X_test)
y_pred = (y_pred_proba > 0.5).astype(int)

# Flatten y_test_np for evaluation functions if it's not already 1D
y_test_flat = y_test_np.flatten()

print("### Model Evaluation ###")

# 1. Classification Report
print("\nClassification Report:")
print(classification_report(y_test_flat, y_pred))

# 2. Confusion Matrix
conf_matrix = confusion_matrix(y_test_flat, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Legitimate (0)', 'Phishing (1)'],
            yticklabels=['Legitimate (0)', 'Phishing (1)'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

# 3. ROC Curve and AUC
fpr, tpr, thresholds = roc_curve(y_test_flat, y_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

Inference: The trained deep learning model achieved 97% accuracy on the test set, with high precision, recall, and F1-scores for both legitimate and phishing classes, indicating strong classification performance.
